## Câu 1: Viết hàm <code>Gauss_elimination(&Atilde)</code>, trong đó:
- **Input**: $\tilde{A}$, là ma trận mở rộng của hệ phương trình $Ax=b$.
- **Output**: ma trận có dạng bậc thang có được từ ma trận $\tilde{A}$.<br>
Sinh viên không được dùng các hàm có sẵn của các thư viện để tìm ma trận bậc thang. 

### Phép khử Gauss (Gaussian elimination):
**Khử Gauss (Gaussian elimination)** là một cách biến đổi tương đương dòng đưa ma trận về dạng bậc thang. Thuật giải gồm các bước:
- **Bước 1**: Xác định cột trái nhất không chứa toàn số $0$.
- **Bước 2**: Nếu đầu cột nhận được ở trên là số $0$, đổi chỗ hai dòng, để đưa số hạng khác $0$ đầu tiên ở dưới về đầu cột.
- **Bước 3**: Với số hạng đầu cột nhận được từ Bước 2 là $a \neq 0$, nhân dòng chứa nó với $\frac{1}{a}$ để có số dẫn đầu $a=1$.
- **Bước 4**: Cộng một bội số thích hợp của dòng đầu cho từng dòng dưới để biến các số dẫn đầu cùng cột bên dưới thành $0$.
- **Bước 5**: Che dòng đầu đã làm xong, tiến tới cột tiếp theo. Lặp lại các bước trên cho đến khi hết cột hoặc không tìm được cột nào không chứa toàn số $0$.

```python
# Hàm khử Gauss (Gaussian-elimination):
def Gauss_elimination(A):
    # Kiểm tra đầu vào
    if not A:
        print("Ma trận rỗng!")
        return A

    nrow, ncol = len(A), len(A[0])
    rootCol, rootRow = 0, 0
    # Lặp với biến "root" coi như là "gốc" xét, hay gọi là hàng/cột đầu tiên
    while rootCol < ncol and rootRow < nrow:
        # B1,2: Nếu phần tử ở đầu cột là $0$, ta kiểm tra các dòng dưới và hoán đổi dòng nếu có phần tử khác $0$
        if A[rootRow][rootCol] == 0:
            for i in range(rootRow, nrow):
                if A[i][rootCol] != 0: # Tìm được phần tử != 0, hoán vị dòng gốc
                    A[rootRow], A[i] = A[i], A[rootRow]
                    break;
           # Nếu cột toàn $0$, ta chuyển sang cột tiếp theo
            if A[rootRow][rootCol] == 0:
                rootCol += 1
                continue
        # B3: Nếu phần tử ở đầu cột khác $1$, ta nhân dòng đó với $\frac{1}{a}$ để đưa phần tử đó về $1$
        if A[rootRow][rootCol] != 1.0:
            A[rootRow] = [1/A[rootRow][rootCol] * x for x in A[rootRow]]
        # B4: Trừ lần lượt một bội số phù hợp của dòng đầu tiên từ các dòng dưới để đưa các phần tử ở dưới dòng đó về $0$,  nếu bằng $0$ sẵn rồi thì bỏ qua
        for i in range(rootRow+1, nrow):
            if -A[i][rootCol] != 0:
                A[i] = [A[i][j] - A[i][rootCol]*A[rootRow][j] for j in range(ncol)]
        # B5: Sau khi hoàn thành các bước trên cho một cột, tiến hành che dòng đã hoàn thành và di chuyển sang cột tiếp theo
        rootRow += 1
        rootCol += 1
    return A
# Main - test-cases:
A = [[1, 2, -1, 5], [2, 3, -2, 6], [3, 4, -3, 7]]
print("Ma trận ban đầu:"); print(A)
Gauss_elimination(A)
print("Ma trận bậc thang: "); print(A)
print("----")
A = [[1, 2, -1, 5], [2, 4, -2, 10], [3, 6, -3, 15]]
print("Ma trận ban đầu:"); print(A)
Gauss_elimination(A)
print("Ma trận bậc thang: "); print(A)
```

## Câu 2: Viết hàm <code>back_substitution(A)</code>, trong đó
- **Input**: $A$, là ma trận có dạng bậc thang thu được từ ma trận mở rộng của hệ phương trình $Ax=b$.
- **Output**: nghiệm của hệ phương trình (trường hợp nghiệm duy nhất/ vô số nghiệm) hoặc thông báo hệ phương trình vô nghiệm.<br>
Sinh viên không được dùng các hàm có sẵn của các thư viện để tìm nghiệm của hệ phương
trình.<br>
Trong quá trình thể hiện nghiệm tổng quát (cho trường hợp vô số nghiệm), sinh viên có thể
sử dụng thư viện để hỗ trợ (lưu ý: chỉ hỗ trợ).

### Phép thế ngược (Backward substitution):
**Phép thế ngược (Backward substituition)** là một phương pháp để giải một hệ phương trình tuyến tính, với đầu vào là một ma trận bậc thang được lấy từ **phép khử Gauss**, và đầu ra là kết quả giải nghiệm của hệ phương trình tuyến tính ban đầu.
Các bước phép thế ngược được trình bày như sau:
- **Bước 1**: Xác định tính có nghiệm của hệ phương trình:
    - Nếu trong ma trận có tồn tại ít nhất một dòng, mà tất cả các phần tử trên dòng đó trừ phần tử cuối cùng, bằng $0$, thì kết luận **Hệ phương trình vô nghiệm** và kết thúc.
    - Ngược lại nếu không tồn tại bất kỳ dòng nào như vậy thì hệ phương trình có nghiệm.
- **Bước 2**: Tính hạng của ma trận: rank(A) $=$ số lượng dòng khác $0$ của ma trận.
- **Bước 3**: Xét:
    - Nếu hạng của ma trận $>=$ số ẩn số của hệ phương trình thì kết luận **Hệ phương trình có nghiệm duy nhất**. 
    - Nếu hàng của ma trận $<$ số ẩn số của hệ phương trình thì kết luận **Hệ phương trình có vô số nghiệm**.<br>
    Với biến $x_i$ và tập các cột chứa hệ số dẫn đầu (khác 0) trên từng dòng ma trận S. Nếu $i\notin S$ thì gán ẩn tự do $x_i=t_k$ là một hằng số $\in \mathbb{R}$, $(k=1,2,3,...)$.
- **Bước 4**: Tiến hành thế ngược bằng cách lặp trên từng dòng từ dưới lên. Giá trị ẩn cần tìm bằng hệ số tự do trừ đi tổng tất cả các số hạng được tạo thành bởi các ẩn đã biết (bao gồm cả ẩn số tự do), tất cả chia cho hệ số dẫn đầu. Tiếp tục sử dụng quá trình này cho đến khi tìm được tất cả các ẩn số. <br>

```python
#### import sympy
from sympy import symbols
# Hàm thế ngược (backward-substitution)
def back_substitution(A):
    # Kiểm tra điều kiện
    if not A:
        print("Ma trận rỗng!")
        return A
    
    # Kiểm tra tính vô nghiệm, đồng thời tính luôn hạng ma trận
    rank = 0
    while rank < len(A):
        if all(x == 0 for x in A[rank][:-1]): # Nếu cả dòng cho đến trước phần tử cuối đều = 0
            if A[rank][-1] == 0: # Nếu cả dòng toàn 0
                break
            else:             # Nếu phần tử cuối != 0
                print("Hệ phương trình vô nghiệm!")
                return []
        rank += 1
            
    # Nếu hệ phương trình có nghiệm:   
    solution = [] # Vector nghiệm
    numberOfVars = len(A[0]) - 1 # Số lượng ẩn số của HPT   
    if rank >= numberOfVars:
        print("Hệ phương trình có nghiệm duy nhất!")
        solution = [0] * numberOfVars
    else:
        print("Hệ phương trình có vô số nghiệm!")
        numberOfFreeVars, rootRow, rootCol = 0, 0, 0
        while rootRow < rank and rootCol < numberOfVars: # A[rootRow][rootCol] là phần tử của đường S
            if A[rootRow][rootCol] == 0: # Nếu gặp 0 thì thêm 1 ẩn tự do, đi sang cột phải
                numberOfFreeVars += 1
                solution.append(symbols("t_" + str(numberOfFreeVars)))
            else: # Nếu gặp khác 0 thì thêm 1 ẩn không tự do, đi chéo xuống (cột, dòng tiếp theo)
                solution.append(0)
                rootRow += 1
            rootCol += 1
        
        # Nếu số phương trình nhỏ hơn số ẩn và các ẩn tự do là những x_cuối
        # Đoạn code sau để khởi tạo đủ ẩn tự do trong vector nghiệm          
        while len(solution) < numberOfVars:
            numberOfFreeVars += 1
            solution.append(symbols("t_" + str(numberOfFreeVars)))

    # Thế ngược từ dưới lên (song song lặp từ x_cuối đến x_đầu, từ dòng cuối đến dòng đầu)            
    i, j = rank - 1, numberOfVars - 1
    while i >= 0 and j >= 0:
        if solution[j] == 0: # Nếu x_j không phải ẩn tự do thì tính dựa theo các ẩn có thứ tự sau nó, sau đó lên dòng trên
            solution[j] = (A[i][-1] - sum(A[i][k]*solution[k] for k in range(j+1, numberOfVars)))/A[i][j]
            i -= 1
        j -= 1 # Nếu x_j là ẩn tự do thì giá trị của nó không thay đổi, đi tính x khác, giữ nguyên dòng hiện tại
    return solution

# Main - test-cases:
A = [[4, -2, 3, 3], [-1, 3, 2, 7], [0, 2, -3, 1]]
print("Ma trận ban đầu:"); print(A)
print("Ma trận bậc thang:"); Gauss_elimination(A); print(A)
solution = back_substitution(A)
if solution: print(f"Vector nghiệm: {solution}")
print("----")
A = [[2, 1, 1, 1], [-1, 2, 4, 4], [3, -4, -2, 3], [4, 1, 15, 13]]
print("Ma trận ban đầu:"); print(A)
print("Ma trận bậc thang:"); Gauss_elimination(A); print(A)
solution = back_substitution(A)
if solution: print(f"Vector nghiệm: {solution}")
print("----")
A = [[1, 2, -1, 5], [2, 4, -2, 10], [3, 6, -3, 15]] 
print("Ma trận ban đầu:"); print(A)
print("Ma trận bậc thang:"); Gauss_elimination(A); print(A)
solution = back_substitution(A)
if solution: print(f"Vector nghiệm: {solution}")
print("----")
```

### Mô tả ý tưởng:
- Hàm thế ngược sẽ tiến hành tương tự giải thuật ở trên.
- Kiểm tra tính vô nghiệm bằng cách: lặp trên từng dòng, tìm xem có tồn tại dòng nào toàn $0$ trừ phần tử cuối cùng là $\neq 0$ hay không.
- Nếu có nghiệm, sử dụng hạng của ma trận để xem HPT có nghiệm duy nhất hay vô số nghiệm:
    - Nếu rank(A) $\geq$ số ẩn thì HPT có nghiệm duy nhất (tính được tất cả nghiệm $x_i$ là giá trị tường minh).
    - Nếu rank(A) $<$ số ẩn thì HPT có vô số nghiệm (phải tìm và đặt ẩn số tự do là $x_i = t_j \in \mathbb{R}$).
- Tìm ẩn tự do (với TH vô số nghiệm): vẽ 1 đường nối (gọi là $S$) bắt đầu từ góc trên bên trái ma trận, nếu:
    - Gặp phần tử khác 0: đi theo đường chéo xuống dòng dưới.
    - Gặp phần tử bằng 0: đi qua phải trên cùng dòng.<br>
    $\Rightarrow$ Ẩn $x_i$ là tự do nếu, $S_i=0$ hoặc $i>\text{[số phần tử của S]}$.<br>
    VD: ta có ma trận sau và vẽ được đường nối: $S=[1,0,1,1]$. 
\begin{bmatrix}
1* & 2 & -2 & 2 & -3 & 3\\
0 & 0* &  1* & 2 & 2 & -1 \\
0 & 0 &  0 & 1* & 2 & -2
\end{bmatrix}
    Khi đó, $x_2$ là ẩn tự do vì $S_2=0$ và $x_5$ là ẩn tự do vì $5 > \text{số phần tử của S} = 4$.
- Sau đó, lặp từ dòng dưới lên trên để tìm giá trị của ẩn từ $x_\text{cuối cùng}$ đến $x_\text{đầu tiên}$ bằng cách lấy giá trị cuối cùng trên từng dòng, trừ đi các số hạng là tích của hệ số với ẩn đã biết, xong chia cho hệ số dẫn đầu hàng hiện tại ($\Rightarrow$ do đó mà ẩn $x_\text{cuối cùng}$ luôn hoặc là ẩn tự do, hoặc được tính trực tiếp là hằng số, sau đó làm tiền đề để giải các ẩn thứ tự trước, cứ như vậy cho đến hết ẩn).

    - Tạo biến <code>solution</code> là vector nghiệm, ban đầu các phần tử của vector nghiệm sẽ được khởi tạo:
        - Là $0$ (nếu là ẩn không tự do).
        - Có dạng $t_k \in \mathbb{R}$ (nếu là ẩn tự do).
    - Tạo biến <code>numberOfVars</code> là $\text{[số cột]}-1$ là số lượng ẩn của HPT.
    - Nếu hạng ma trận không nhỏ hơn số ẩn thì ta có đủ số phương trình giải nghiệm, do đó thông báo **HPT có nghiệm duy nhất**.<br>
    Tiến hành khởi tạo <code>numberOfVars</code> phần tử $0$ trong vector nghiệm <code>solution</code>.
    - Nếu hạng ma trận nhỏ hơn số ẩn thì ta không đủ số phương trình giải nghiệm, do đó thông báo **HPT có vô số nghiệm**. <br>
    Tiến hành như ý tưởng trên mục **Mô tả ý tưởng** (duyệt theo đường $S$) để khởi tạo giá trị cho vector nghiệm <code>solution</code>:
        - Vẽ đường $S$ bằng cách sử dụng kĩ thuật như trong hàm <code>Gauss_elimination(A)</code> bên trên: dùng 1 vòng lặp <code>while</code> với $2$ biến đếm lưu lại vết của dòng, cột là <code>rootRow</code>, <code>rootCol</code> (có thể đọc lại ý tưởng phần hàm khử Gauss bên trên để rõ hơn).
        - Nếu trên đường $S$ gặp phần tử $0$ thì thêm vào vector nghiệm một giá trị dạng <code>t_numberOfFreeVars</code> với biến <code>numberOfFreeVars</code> để đếm số lượng ẩn tự do. Còn không thì thêm giá trị $0$ vào vector nghiệm để đánh dấu một ẩn không tự do.
    - Sau cùng, **tiến hành thế ngược** để hoàn thiện giá trị trong vector nghiệm <code>solution</code>:
        - Tạo hai vòng lặp song song, một vòng lặp <code>i</code> để duyệt hàng từ dưới lên, vòng lặp <code>j</code> để duyệt từ ẩn thứ tự cuối cùng đến ẩn thứ tự đầu tiên trong vector nghiệm <code>solution</code>.
        - Nếu ẩn thứ <code>j</code> là $0$ thì nó không phải ẩn tự do, thì tính nó: lấy giá trị cuối cùng của hàng trừ đi tổng các số hạng (là các tích của hệ số với ẩn đã tính), rồi chia cho hệ số dẫn đầu của dòng hiện tại. Sau khi tính xong thì xét tiếp dòng bên trên và ẩn tiếp theo.
        - Nếu ẩn thứ <code>j</code> khác $0$ thì chắc chắn nó là ẩn tự do (vì đã được tạo trước đó rồi), bỏ qua nó xét tiếp ẩn tiếp, nhưng vẫn giữ nguyên dòng đang xét vì ta chưa khai thác dòng đó.
        - Cứ lặp cho đến khi hết dòng hoặc tính hết ẩn.

### Câu 3:  Mở rộng:
Tìm hiểu hàm/ phương thức tương ứng của các thư viện và thực hiện nó, so sánh kết quả

#### Biến đổi ma trận mở rộng về dạng bậc thang:
- Thư viện <code>sympy</code>, phương thức <code>sympy.Matrix.rref()</code>: biến đổi về ma trận bậc thang rút gọn (Gauss-Jorrdan).
- Thư viện <code>numpy, scipy</code>, phương thức <code>scipy.linalg.lu</code>: phân tích LU để đưa về ma trận bậc thang, với hệ số dẫn đầu chưa chắc bằng $1$.<br>
Đồ án này sẽ cài đặt thử hàm thứ nhất.

#### Giải nghiệm hệ phương trình tuyến tính:
- Thư viện <code>sympy</code>, hàm <code>sympy.linsolve()</code>: có thể nhận đầu vào là ma trận mở rộng, hoặc ma trận hệ số và vector kết quả.
- Thư viện <code>numpy</code>, hàm <code>numpy.linalg.solve()</code>: chỉ nhận đầu vào là ma trận hệ số và vector kết quả.<br>
Đồ án này sẽ cài đặt thử hàm thứ nhất.

```python
import sympy as sp
from sympy import Matrix, linsolve

# Main - test-cases
m = sp.Matrix([[3, -6, -2, 1], [2, -4, 2, 1], [4, -7, -4, 3]])
print("Ma trận ban đầu:"); print(m)
m, pivot = m.rref()
print("Ma trận bậc thang rút gọn: "); print(m)
solution = linsolve(m)
if solution:
    print(f"Hệ phương trình có nghiệm: {solution}")
else:
    print("Hệ phương trình vô nghiệm!")
print("----")
m = sp.Matrix([[2, -4, 6, -6], [4, 4, 0, 0], [0, -6, 8, 2], [2, 0, 2, -2]])
print("Ma trận ban đầuu:"); print(m)
m, pivot = m.rref()
print("Ma trận bậc thang rút gọn: "); print(m)
solution = linsolve(m)
if solution:
    print(f"Hệ phương trình có nghiệm: {solution}")
else:
    print("Hệ phương trình vô nghiệm!")
print("----")
m = sp.Matrix([[3, -1, 1, -3, 0], [4, -1, 4, -2, 0]])
print("Ma trận ban đầu:"); print(m)
m, pivot = m.rref()
print("Ma trận bậc thang rút gọn: "); print(m)
solution = linsolve(m)
if solution:
    print(f"Hệ phương trình có nghiệm: {solution}")
else:
    print("Hệ phương trình vô nghiệm!")
```

#### Rút ra kết luận:
- Phương thức <code>sympy.Matrix.rref()</code>:
    - Đưa ma trận thành dạng bậc thang rút gọn, trong khi hàm <code>Gauss_elimination()</code> chỉ đưa về dạng bậc thang có hệ số dẫn đầu là $1$.
    - Trả về $2$ giá trị là ma trận sau khi rút gọn và vector các cột cơ sở.
    - Có hỗ trợ cả phân số (số hữu tỉ).
- Hàm <code>sympy.linsolve()</code>:
    - Có kết quả trùng với hàm thế ngược <code>back_substitution()</code>.
    - Có thể nhận tham số là một ma trận mở rộng hoặc (ma trận hệ số và vector giá trị).
    - Trả về vector nghiệm, ẩn tự do mặc định đặt tên dạng <code>tau_i</code>, có hỗ trợ phân số cho đẹp.
